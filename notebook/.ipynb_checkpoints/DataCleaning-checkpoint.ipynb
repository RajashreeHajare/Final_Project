```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv("Telco-Customer-Churn.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data.rename(columns = {"gender":"Gender", "tenure": "Tenure","customerID":"CustomerID"}, inplace =True)
```

```python
data.isnull().sum()
```

```python
data.dropna()
```

```python
data.drop_duplicates()
```

```python
data.nunique()
```

```python
data.dtypes
```

```python
data['SeniorCitizen'] = data['SeniorCitizen'].map({1: 'yes', 0: 'no'})
```

```python
data['TotalCharges'] = pd.to_numeric(data['TotalCharges'], errors='coerce')
```

```python
#replace blanks with median 

data['TotalCharges'].fillna(data['TotalCharges'].median(), inplace=True)
```
